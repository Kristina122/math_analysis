### Задание № 3

```python
### график окружности
# x**2  + y**2 = r**2
r = 6
x = np.linspace(-r,r,1000)
y = np.sqrt(-x**2+r**2)
plt.plot(x, y,'b')
plt.plot(x,-y,'b')
plt.gca().set_aspect('equal') ## уравнивание масштаба осей
plt.show()
```

```python
### график эллипса
# (x/a)**2 + (y/b)**2 = 1
a = 10
b = 5
x = np.linspace(-a,a,1000)
y = np.sqrt(b**2 - ((x**2*b**2)/a**2))
plt.plot(x, y,'b')
plt.plot(x,-y,'b')
plt.gca().set_aspect('equal') ## уравнивание масштаба осей
plt.show()
```

```python
## график гиперболы
# (x/a)**2 - (y/b)**2 = 1
a = 5
b = 5
x = np.linspace(a,a*2,10000)
ytop = b*np.sqrt((x**2/a**2) - 1)
ybot = -b*np.sqrt((x**2/a**2) - 1)
# y = np.sqrt(x**2-1)
plt.plot(-x, ytop,'b')
plt.plot(x, ytop,'b')
plt.plot(x,ybot,'b')
plt.plot(-x,ybot,'b')
plt.gca().set_aspect('equal') ## уравнивание масштаба осей
plt.show()
```

###### 4.1

***"Если плоскости A1*x + B1*x + C1*z + D1=0 и A2*x + B2*y + C2*z + D2 = 0 параллельны, то нормальные векторы N1(A1,B1,C1) N2(A2, B2, C2) коллинеарны (и обратно). Поэтому условие A2/A1 = B2/B1 = C2/C1 есть необходимое и достаточное условие параллельности или совпадения плоскостей."*** (инф. из Википедии)

=> уравнение плоскости параллельной **A*x + B*y + C*z + D = 0** и проходящей через начало координат будет выглядить слудеющим образом: 
$$A1*x + B1*y + C1*z + 0 = 0$$ при условии, что **A1/A = B1/B = C1/C.**
